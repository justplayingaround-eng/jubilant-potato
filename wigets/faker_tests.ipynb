```python
import pandas as pd
from faker import Faker
import random
from datetime import timedelta

fake = Faker()
Faker.seed(100)
# 1. Setup Constraints
NUM_UNIQUE_NAMES= 10
TOTAL_ROWS = 3000
START_DATE = pd.Timestamp("2024-06-01")
END_DATE = pd.Timestamp("2024-06-30")

# 2. Create the "Source of Truth" 
name_city_registry = []
for _ in range(NUM_UNIQUE_NAMES):
    name_city_registry.append({
        "name": fake.name(),
        "city": fake.city(),
        "amount":  random.randint(0,1000)
    })

# 3. Generate the 300 log rows
log_data = []
for x in range(TOTAL_ROWS):
    # Select a random NAME/City pair from our registry
    random.seed(x)
    identity = random.choice(name_city_registry)
    
    seen = fake.date_time_between(start_date=START_DATE, end_date=END_DATE)
    
    log_data.append({
        "name": identity["name"],
        "city": identity["city"],
        "seen": seen,
        "amount": identity["amount"],
    })

# 4. Convert to DataFrame and sort by time
df = pd.DataFrame(log_data).sort_values(by="seen").reset_index(drop=True)

df
```
