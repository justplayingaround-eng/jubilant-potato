```python
from ipywidgets import AppLayout, Button, Layout
```

```python
# Centering the text content itself
title = widgets.HTML(
    value="<div style='text-align: center; width: 100%;'><h1>My Dashboard Title</h1></div>"
)

StartDatetimePicker = widgets.NaiveDatetimePicker(description='Pick a Time')
EndDatetimePicker = widgets.NaiveDatetimePicker(description='Pick a Time')

RadioButtons = widgets.RadioButtons(
    options=['pepperoni', 'pineapple', 'anchovies'],
#    value='pineapple', # Defaults to 'pineapple'
#    layout={'width': 'max-content'}, # If the items' names are long,
    disabled=False, layout=widgets.Layout(justify_items='center',align_items='center')
)

Select = widgets.Select(
    options=['Linux', 'Windows', 'macOS'],
    value='macOS',
    # rows=10,
    description='OS:',
    disabled=False, layout=widgets.Layout(justify_items='center',align_items='center')
)

user_options = HBox([VBox([widgets.HTML(
    value="<div style='text-align: center; width: 100%;'><h3>Plot</h3></div>"
),RadioButtons]),VBox([widgets.HTML(
    value="<div style='text-align: center; width: 100%;'><h3>Plot</h3></div>"
),StartDatetimePicker,EndDatetimePicker]),VBox([widgets.HTML(
    value="<div style='text-align: center; width: 100%;'><h3>Plot</h3></div>"
),Select])])

accordion = widgets.Accordion(children=[ widgets.Text()], titles=('Text',))

user_options_with_filters = VBox([user_options,accordion])


go_button = Button(description='Go', button_style='success', layout=Layout(height='50px', width='auto'))


AppLayout(header=title,
          left_sidebar=None,
          center=user_options_with_filters,
          right_sidebar=None,
          footer=go_button,layout=Layout(border = '3px solid black'))
```
